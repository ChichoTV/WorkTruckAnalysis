```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
# test url of pickup trucks filtered by zip code and mile radius
# url='https://www.comvoy.com/vehicles/pickup-trucks-for-sale-1fa7?page={pg}&filters=Area:Zip:{zip};Miles:{mi}'
url='https://www.comvoy.com/vehicles/pickup-trucks-for-sale-1fa7?page={pg}&filters=Area:Zip:{zip};Miles:{mi}'
html=requests.get(url.format(pg=1,zip=95973,mi=10000)).text
soup=BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
prices=soup.body.find_all('span', {'class': 'data value'})
articles=soup.body.find_all('article')
# finding price by article-------float(articles[0].find_all('span', {'class': 'data value'})[0]['content'])
```

```python
# example of data value-----data_table[0].dl.div.dd.text
# column label can be found this way------data_table[0].dl.div.dt.text
```

```python
data_array=[]
while True:
    try:
        i=1
        for table in articles:
            mydict=dict()
            mydict['price']=float(table.find_all('span', {'class': 'data value'})[0]['content'])
            for div in table.find_all("div", {"class": "truck__specifications specifications"})[0].find_all('div'):
#             print(div)
            mydict[div.dt.text]=div.dd.text
            data_array.append(mydict)
```
